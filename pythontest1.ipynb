```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv('Iris.data.csv', header = None, names=names)
from pandas import DataFrame
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
iris = iris.drop('class', axis = 1)
pca = PCA(n_components=2).fit(iris)
df2 = pca.transform(iris)
kmeans = KMeans(n_clusters = 3).fit(df2)
centroids = kmeans.cluster_centers_
for i in range (0, df2.shape[0]):
  if kmeans.labels_[i] == 0:
    plt.scatter(df2[i,0], df2[i,1], c='green')
  elif kmeans.labels_[i]==1:
    plt.scatter(df2[i,0], df2[i,1], c='blue')
  elif kmeans.labels_[i] == 2:
    plt.scatter(df2[i,0], df2[i,1], c='yellow')
plt.scatter(centroids[:,0], centroids[:,1],c ='black', marker ='X')
plt.title('Clustering')
```
